```python
#importing libraries
import numpy as np        #data handling
import pandas as pd       #visualization
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt #visualization
%matplotlib inline
#use matplotlib inside the notebook
```

```python
url='https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
col_name=['sepal-length','sepal-width','petal-length','petal-width','class']
dataset=pd.read_csv(url,names=col_name)
```

```python
#Pandas. provides a function called head() to print the top five rows of the datset
dataset.shape
```

```python
dataset.head()
```

```python
#Pandas provides a function called describe() to compute and print out summary statistics for each class, such as the number of data points, mean value, standard deviation, minimum value, maximum value and the quartiles.
dataset.describe()
```

```python
#Pandas provides a function called info() that prints out the data type of each column andthe memory usage.
dataset.info()
```

```python
#Pandas provides a function called value_counts() to show us the number of examples in each class.
dataset['class'].value_counts()
```

```python
# Violin Plot
sns.violinplot(y='class', x='sepal-length', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='sepal-width', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='petal-length', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='petal-width', data=dataset, inner='quartile')
plt.show()
```

```python
# Violin Plot
sns.violinplot(y='class', x='sepal-length', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='sepal-width', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='petal-length', data=dataset, inner='quartile')
plt.show()
sns.violinplot(y='class', x='petal-width', data=dataset, inner='quartile')
plt.show()
```

```python
# Violin Plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # Create a 2x2 grid of plots

# Plot 1: Sepal Length vs Class
sns.violinplot(y='class', x='sepal-length', data=dataset, inner='quartile', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length vs Class')

# Plot 2: Sepal Width vs Class
sns.violinplot(y='class', x='sepal-width', data=dataset, inner='quartile', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width vs Class')

# Plot 3: Petal Length vs Class
sns.violinplot(y='class', x='petal-length', data=dataset, inner='quartile', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length vs Class')

# Plot 4: Petal Width vs Class
sns.violinplot(y='class', x='petal-width', data=dataset, inner='quartile', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width vs Class')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Pair Plot
sns.pairplot(dataset, hue='class', markers='+')
plt.show()
```

```python
# Select only numerical features for correlation analysis
numerical_features = dataset.select_dtypes(include=['number'])

#correlation heatmap
plt.figure(figsize=(7,5))
sns.heatmap(numerical_features.corr(), annot=True, cmap='jet')
plt.show()
```
